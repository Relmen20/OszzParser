```python
def take_text(tov_title_element):
    my_dict_article = dict()
    chain = []
    art_element_txt="no article"
    
    try:
        art_elements = tov_title_element.find_elements(By.TAG_NAME, 'a')
        for element in art_elements: 
            if len(element.text) > 2:
                art_element_txt = element.text
    except:
        art_element = tov_title_element.find_element(By.TAG_NAME, 'a')
        art_element_txt = art_element.text
        
    manufacture_element = tov_title_element.find_element(By.TAG_NAME, 'strong')
    description_element = tov_title_element.find_element(By.CLASS_NAME, 'text-muted')
    chain.append(manufacture_element.text)
    chain.append(description_element.text)
    
    
    print(art_element_txt)
    
    my_dict_article[art_element_txt] = chain
    print(my_dict_article)
    return my_dict_article
```

```python
def result(driver, id, result_element):
    # global driver, id, my_dict_id, my_dict_article, result_element
    requested_list = []
    chain = []
    my_dict_article = dict()
    my_dict_id = dict()
    
    if (id == "Requested"):
        try:
            requested_element = result_element.find_element(By.ID, id) 
            tov_title_element = requested_element.find_element(By.CLASS_NAME, 'tov-title')
            
            my_dict_article = take_text(tov_title_element)
            if my_dict_article: 
                requested_list.append(my_dict_article)
            
            if requested_list: 
                my_dict_id[id] = requested_list
                
        except Exception as e:
            print("", end="")   
            
    else:
        try:
            requested_element = result_element.find_element(By.ID, id)
            link_elements = requested_element.find_elements(By.TAG_NAME, 'a')
            for link in link_elements:
                link.click()
                time.sleep(1)

            tov_title_elements = requested_element.find_elements(By.CLASS_NAME, 'tov-title')
            for tov_title_element in tov_title_elements:
                time.sleep(1)
                
                my_dict_article = take_text(tov_title_element)
                
                if my_dict_article: 
                    requested_list.append(my_dict_article)
                
            if requested_list: 
                my_dict_id[id] = requested_list    
                
        except Exception as e:
            print("", end="")      
            
    return my_dict_id
```

```python
def change_proxy(driver):
    global PROXY_NUMBER, free_proxies
    
    if PROXY_NUMBER == 0:
        free_proxies = get_free_proxies()
    proxy = free_proxies[PROXY_NUMBER]
    print(proxy)
    driver.close()
    driver.quit
    chrome_options = uc.ChromeOptions()
    chrome_options.add_argument(f'--proxy-server={proxy}')
    chrome_options.add_argument('--window-size=1024,768')
    chrome_options.add_argument("--incognito")
    driver = uc.Chrome(options=chrome_options)
    
    PROXY_NUMBER+=1
    if PROXY_NUMBER == len(free_proxies):
        PROXY_NUMBER = 0
        
    try:
#         driver.set_page_load_timeout(10)
        driver.get("https://exist.ru/")
        time.sleep(10)
        driver.close()
        driver.quit
    except:
        driver = change_proxy(driver)
    return driver
        
def get_free_proxies():
    time.sleep(1)
    try:
        response = requests.get(url)
        response.raise_for_status()  
        data = response.content
        lines = data.split(b'\n')
        proxies = [line.decode('utf-8') for line in lines if line]

    except requests.exceptions.RequestException as e:
        print(f"An error occurred during the request: {str(e)}") 
    except requests.exceptions.HTTPError as e:
        print(f"API returned an error: {str(e)}")  
    except ValueError as e:
        print(f"Failed to parse JSON: {str(e)}")
    return proxies
```

<div class="alert alert-success"><div>Def that filtered TXT file out of non-article trash</div></div>

```python
def contains_letters(line):
    return any(char.isalpha() for char in line)

def filter_lines(input_filename, output_filename):
    try:
        with open(input_filename, 'r') as input_file:
            lines = input_file.readlines()

        filtered_lines = [line for line in lines if contains_letters(line)]

        with open(output_filename, 'w') as output_file:
            output_file.writelines(filtered_lines)

        print("Filtered lines written to", output_filename)
    except FileNotFoundError:
        print("Input file not found.")

# Replace these with your actual file paths
input_file_path = 'articles.txt'
output_file_path = 'output.txt'

filter_lines(input_file_path, output_file_path)

with open(output_file_path, 'r') as input_file:
            lines = input_file.readlines()
index = 1
for line in lines:
    print(f"{index}) {line}", end='')
    index+=1
```

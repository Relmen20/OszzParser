```python
from openpyxl import Workbook
from openpyxl.styles import Alignment, Font, Border, Side
import pickle

with open('copied_data.pkl', 'rb') as file:
    data = pickle.load(file)
    
path = "clear_table_only_changable_OSZZ.xlsx"
wb = Workbook()
sheet1 = wb.active
alignment = Alignment(horizontal="center", vertical="center")
font = Font(size=14, bold=True)
sheet1.column_dimensions['A'].width = 45/10
sheet1.column_dimensions['B'].width = 220/10
sheet1.column_dimensions['C'].width = 300/10
sheet1.column_dimensions['D'].width = 350/10
sheet1.column_dimensions['E'].width = 200/10
sheet1.column_dimensions['F'].width = 610/10
sheet1.row_dimensions['1'].height = 25
for i in range(1,7):
    sheet1.cell(row=1, column=i).alignment = alignment
    sheet1.cell(row=1, column=i).font = font

sheet1.cell(row=1, column=1).value = '№'
sheet1.cell(row=1, column=2).value = 'Артикул'
sheet1.cell(row=1, column=3).value = 'Оригинальная замена'
sheet1.cell(row=1, column=4).value = 'Не оригинальная замена'
sheet1.cell(row=1, column=5).value = 'Производители'
sheet1.cell(row=1, column=6).value = 'Название'

index = 1
for key,value in data.items():
    if len(value) >= 2:
        index+=1
        sheet1.cell(row=index, column=1).value = index -1
        sheet1.cell(row=index, column=1).alignment = alignment
        sheet1.cell(row=index, column=2).value = key
        sheet1.cell(row=index, column=2).alignment = alignment
        factory = []
        for item in value:
            for key_id,value_id in item.items():
                ids = []
                if key_id == "Requested":
                    for item2 in value_id:
                        for key_id_art,value_id_art in item2.items():
                            sheet1.cell(row=index, column=6).value = value_id_art[1]
                            factory.append(value_id_art[0])
                            
                if key_id == "OriginalReplacements":
                    for item2 in value_id:
                        for key_id_art,value_id_art in item2.items():
                            ids.append(key_id_art)
                            factory.append(value_id_art[0])
                    data_string = '; '.join(map(str, ids))
                    sheet1.cell(row=index, column=3).value = data_string
                    
                if key_id == "NonOriginalReplacements":
                    for item2 in value_id:
                        for key_id_art,value_id_art in item2.items():
                            ids.append(key_id_art)
                            factory.append(value_id_art[0])
                    data_string = '; '.join(map(str, ids))
                    sheet1.cell(row=index, column=4).value = data_string 
        data_string = '; '.join(map(str, list(set(factory))))            
        sheet1.cell(row=index, column=5).value = data_string


wb.save(path)
```
